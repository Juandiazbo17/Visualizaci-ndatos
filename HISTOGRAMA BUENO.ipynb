```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
%matplotlib inline
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 100)

path = "Autobus.vs.Metro.csv"
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
Autobus=df[df["TRANSPORTE"]==2]
```

```python
plt.figure(figsize=(25,20))

plt.subplot(1,2,1)
plt.scatter(x = Autobus["Kilometros recorridos"], y = Autobus["Pasajeros"])
plt.xlabel("Km")
plt.ylabel("Pasajeros")
plt.title("Diagrama de dispersión de Km vs. Pasajeros Autobus")
```

```python
Metro=df[df["TRANSPORTE"]==3]
```

```python
plt.figure(figsize=(25,20))

plt.subplot(1,2,1)
plt.scatter(x = Metro["Kilometros recorridos"], y = Metro["Pasajeros"])
plt.xlabel("Km")
plt.ylabel("Pasajeros")
plt.title("Diagrama de dispersión de Km vs. Pasajeros Metro")
```
